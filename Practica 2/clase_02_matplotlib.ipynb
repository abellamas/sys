# Graficos y Matplotlib (intro)

Como podrán imaginar, existen numerosas bibliotecas en Python para graficar señales. Todas tienen sus pro y sus contras, y de acuerdo a la aplicación en la que estemos trabajando vamos a elegir una por sobre otra. En esta cátedra utilizamos la libreria Matplotlib, particularmente el modulo PyPlot. Esta es una biblióteca que fue creada con la intención de replicar la herramienta de graficos que ofrece MATLAB. 
Como en toda biblioteca nueva, lo mas recomendable para aprender y empezar a usarla es darle un vistazo a la [documentacion oficial](https://matplotlib.org/3.5.1/tutorials/introductory/usage.html), que en este caso es muy amena y clara, con numerosos ejemplos y comentarios que son de mucha utilidad. Es clave darle importancia al uso y manejo de los gráficos, ya que por lo general son las herramientas que más utilizamos para comunicar nuestros resultados o para realizar análisis.

Existen dos maneras principales de utilizar Matplotlib para graficar:

   - A traves de la [API de pyplot](https://matplotlib.org/3.5.1/tutorials/introductory/pyplot.html)
   - A traves de la [API orientada a objetos](https://matplotlib.org/3.5.1/api/index.html)
    
*(Nota: API --> Interfaz de Programacion de Aplicaciones)*


La API de pyplot es mas sencilla de utilizar, pero en general es menos flexible que la API orientada a objetos, ademas de que la mayoria de los ejemplos de la documentación utilizan esta última, a exepcion de la [documentacion de pyplot](https://matplotlib.org/3.5.1/tutorials/introductory/pyplot.html) (que recomiendo fuertemente revisar antes de comenzar a plotear!). En general, todas las acciones y opciones para graficar que vamos a necesitar se pueden conseguir de las dos maneras, pero sus implementaciones no son iguales. 
**La recomendacion en este punto es empezar por utilizar la API de pyplot para comprender las bases de la biblioteca, y luego pasar de a poco a la utilización de la API orientada a objetos a medida que se van incorporando conceptos.**

Para ver que pinta tienen ambas formas de usar matplotlib, veamos un ejemplo de graficos con cada una


```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,1, 16000)

seno = np.sin(2*np.pi*3*t)
coseno = np.cos(2*np.pi*3*t)
```

### Ejemplo inicial: API de Pyplot

```python
plt.plot(t, seno)
plt.plot(t, coseno)

plt.title("Grafico de ejemplo", fontsize=15)
plt.xlabel("Tiempo [s]", fontsize=13)
plt.ylabel("Amplitud", fontsize=13)
plt.grid()

plt.xlim(0,1)
plt.ylim(-2,2)

#IMPORTANTE en scripts:
plt.show()
```

### Ejemplo inicial: API orientada a objetos

```python
fig, ax = plt.subplots()

ax.plot(t, seno)
ax.plot(t, coseno)
ax.set_title("Ejemplo API orientada a objetos", fontsize=15)
ax.set_xlabel("Tiempo [s]", fontsize=13)
ax.set_ylabel("Amplitud", fontsize=13)
ax.grid()
plt.xlim(0,1)
plt.ylim(-2,2)

#IMPORTANTE en scripts:
plt.show()

```

### Subplots


```python
# pyplot API
plt.figure(1,figsize=(20,5))

plt.subplot(1,2,1)
plt.title('Seno')
plt.plot(t,seno)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid()

plt.subplot(1,2,2)
plt.plot(t,coseno, "r")
plt.title('Coseno')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid()
```

```python
# API orientada a objetos

fig, axs = plt.subplots(1,2,figsize=(20,5))
axs[0].set_title('Seno')
axs[0].plot(t, seno)
axs[0].set_xlabel('Tiempo [s]')
axs[0].set_ylabel('Amplitud')
axs[0].grid()

axs[1].set_title('Coseno')
axs[1].plot(t, coseno, "r")
axs[1].set_xlabel('Tiempo [s]')
axs[1].set_ylabel('Amplitud')
axs[1].grid()
```

```python
funciones = [seno, coseno, seno**3, np.abs(seno)]
nombres = ["Seno", "Coseno", "Seno al cubo", "Valor absouto del seno"]
colores = ["b", "r", "black", "orange"]

# Ejemplo de simplificacion
fig, axs = plt.subplots(4,1,figsize=(10,15))


for i in range(len(funciones)):
    axs[i].set_title(nombres[i], fontsize=15)
    axs[i].plot(t, funciones[i], colores[i])
    axs[i].set_xlabel('Tiempo [s]', fontsize=13)
    axs[i].set_ylabel('Amplitud', fontsize=13)
    axs[i].grid()

fig.tight_layout()
```
